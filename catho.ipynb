```python
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# 'press_cookie' serve para lidar com pop-ups de cookies: 
#  Ela espera até que um botão específico seja clicável e, em seguida, clica nele para aceitar os cookies. 
#  Se não for encontrado, exibe uma mensagem de "Cookie not found".

def press_cookie(driver):
    try:
        botao_cookie = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="lgpd-consent-widget"]/section/div/div[2]/button[1]'))  # Change the selector to match your website
        )
        botao_cookie.click()
    except:
        print('Cookie not found')


# criando uma função para agrupar os 4 elementos do dicionário:
# extract_element recebe um dicionário dicionario_opcoes, um objeto driver do Selenium e um campo campo. 
# Ela itera sobre os seletores CSS presentes no dicionário para encontrar elementos correspondentes no site. 
# Retorna uma lista de elementos que correspondem ao campo especificado.

def extract_element(dicionario_opcoes, driver, campo):
        for i in dicionario_opcoes[campo]:
            lista_elementos = driver.find_elements(By.CSS_SELECTOR, i)
        # print(titulos)
            if len(lista_elementos) > 0:
                break

        return lista_elementos




def extract_info(driver, dicionario_opcoes):
    # criando um dicionario 
    elementos_scrapping = {
    'titulo': [],
    'empresa': [],
    'salario': [],
    'local':[]
    }

    for i in ['titulo', 'empresa', 'salario', 'local']:
        elementos_scrapping[i] = extract_element(dicionario_opcoes, driver, i)

    # retornar listas:
    todos_titulos = [x.text for x in elementos_scrapping['titulo']]
    urls = [x.get_attribute('href') for x in elementos_scrapping['titulo']]
    todas_empresas = [x.text for x in elementos_scrapping['empresa']]
    todas_salario = [x.text for x in elementos_scrapping['salario']]
    todos_local = [x.text for x in elementos_scrapping['local']]

    for i in ['titulo', 'empresa', 'salario', 'local']: # para testar se todas as colunas possuem 15 elementos
        print(len(i))

    df = pd.DataFrame({'titulos': todos_titulos, 'empresa': todas_empresas, 'todas_salario': todas_salario, 'todos_local': todos_local, 'url': urls})
    return df
```
